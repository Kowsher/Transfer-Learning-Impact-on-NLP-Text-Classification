```python
import matplotlib.pyplot as plt

# Data for the accuracy results
models = ['BERT', 'DistilBERT', 'ELECTRA', 'RoBERTa', 'ALBERT']
accuracy_sst2 = [0.923, 0.907, 0.945, 0.925, 0.807]
accuracy_cola = [0.812, 0.785, 0.844, 0.821, 0.897]
accuracy_emotion = [0.802, 0.785, 0.769, 0.808, 0.767]

# Create a figure and axis
plt.figure(figsize=(8, 6))
ax = plt.gca()

# Plot the accuracy results for each dataset with shadow
ax.plot(models, accuracy_sst2, marker='o', label='SST-2', linewidth=2, markersize=8)
ax.plot(models, accuracy_cola, marker='o', label='CoLA', linewidth=2, markersize=8)
ax.plot(models, accuracy_emotion, marker='o', label='Emotion', linewidth=2, markersize=8)

# Highlight the maximum accuracy for each dataset
ax.annotate(f"Max: {max(accuracy_sst2):.3f}", xy=(models[accuracy_sst2.index(max(accuracy_sst2))], max(accuracy_sst2)),
            xytext=(10, 10), textcoords='offset points', arrowprops=dict(arrowstyle="->"))
ax.annotate(f"Max: {max(accuracy_cola):.3f}", xy=(models[accuracy_cola.index(max(accuracy_cola))], max(accuracy_cola)),
            xytext=(-40, 15), textcoords='offset points', arrowprops=dict(arrowstyle="->"))
ax.annotate(f"Max: {max(accuracy_emotion):.3f}", xy=(models[accuracy_emotion.index(max(accuracy_emotion))], max(accuracy_emotion)),
            xytext=(-40, 15), textcoords='offset points', arrowprops=dict(arrowstyle="->"))

# Set labels and title
ax.set_xlabel('Models')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy Comparison for Different Datasets')
plt.xticks(rotation=45)
ax.legend()

# Add a subtle grid
ax.grid(alpha=0.3)

# Show the plot
plt.tight_layout()
plt.show()
```
